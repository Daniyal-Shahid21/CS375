```python
import numpy as np

class Perceptron(object):
    """Perceptron classifier with early stopping and weight history."""

    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
        # ------ New Block: track actual epochs ------
        self.n_iter_run_ = 0
        # ------ End of New Block ------

    def fit(self, X, y):
        """Fit training data."""
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.errors_ = []
        # ------ New Block: maintain weight history after each sample processed ------
        self.w_history_ = []
        # ------ End of New Block ------

        for epoch in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                # ------ Fixed Code: incorrect update sign ------
                # update = self.eta * self.predict(xi) - target
                # ------ End of Fixed Code ------

                # ------ New Block: correct perceptron update ------
                yHat = self.predict(xi)
                update = self.eta * (target - yHat)
                # ------ End of New Block ------

                self.w_[1:] += update * xi
                self.w_[0] += update

                # ------ New Block: log weights after processing this sample ------
                self.w_history_.append(self.w_.copy())
                # ------ End of New Block ------

                errors += int(update != 0.0)

            self.errors_.append(errors)

            # ------ New Block: early stopping when converged ------
            if errors == 0:
                self.n_iter_run_ = epoch + 1
                break
            # ------ End of New Block ------
        else:
            # ------ New Block: if no early stop, record full epochs ------
            self.n_iter_run_ = self.n_iter
            # ------ End of New Block ------

        # ###### New code for doing nothing. - MEH
        this_code_does_nothing = True
        # ######

        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
from matplotlib.colors import ListedColormap


def plot_decision_regions(X, y, classifier, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'o', 'P', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class examples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0],
                    y=X[y == cl, 1],
                    alpha=0.8,
                    c=colors[idx],
                    marker=markers[idx],
                    label=cl,
                    edgecolor='black')
```

```python
plot_decision_regions(X, y, classifier=ppn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')


# plt.savefig('images/02_08.png', dpi=300)
plt.show()
```
